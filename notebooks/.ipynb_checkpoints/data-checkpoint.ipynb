# Model Predictions

```python
model_name='halo_cnn1d_r'
```

```python
data_file = os.path.join(wdir, 'saved_models', model_name, 'model_data',
                       model_name + '_123' + '.npy')
sdm_file = os.path.join(wdir,'raw','UM_0.5_MLv_0_preds.npy')
```

```python
out = np.load(data_file, 
                encoding='latin1').item()
par = out['params']
sdm_dat = matt.parseout(np.load(sdm_file))
```

```python
print('Power law predictions\n')

y_regr = np.reshape(out['y_regr'], (len(out['y_regr']),1))
y_regr_test = np.reshape(out['y_test'], (len(out['y_test']),1))


regr = linear_model.LinearRegression(fit_intercept=True)
regr.fit(y_regr, out['sigv_regr'])

print('regr coef: ' +  str(regr.coef_))
print('regr intercept: ' + str(regr.intercept_))
print('regr R^2: ' + str(regr.score(y_regr,out['sigv_regr'])))

y_regr_pred = (out['sigv_test'] - regr.intercept_)/regr.coef_
y_regr_pred = y_regr_pred.flatten()
```

```python
one_to_one = np.arange(11)*(par['Y_max'] - par['Y_min'])/10. + par['Y_min']

f = plt.figure(figsize=[4,6])
gs = mpl.gridspec.GridSpec(2,1,height_ratios=[1,1], hspace=0)

ax1 = f.add_subplot(gs[0,0])

fit_y = (one_to_one * regr.coef_ + regr.intercept_).flatten()
ax1.plot(one_to_one,fit_y,'r',label='fit')

matt.binnedplot(out['y_regr'],
                 out['sigv_regr'],
                 n=50,
                 percentiles = [35,47.5],
                 ax=ax1,
                 label='train_',
                 names=True,
                 c='g',
                 log=0
                )
ax1.set_xlim(xmin=par['Y_min'], xmax=par['Y_max'])

ax1.set_ylabel('$\log[\sigma_v]$', fontsize=14)
ax1.legend(fontsize=8,loc=4)

ax2 = f.add_subplot(gs[1,0],sharex=ax1)
                
ax2.plot(one_to_one,one_to_one,'k',linestyle='dashed')
matt.binnedplot(out['y_test'],
                 y_regr_pred,
                 n=75,
                 percentiles = [35],
                 ax=ax2,
                 label='test_',
                 names=True,
                 log=0
                )


ax2.set_xlabel('$\log[M$]', fontsize=14)
ax2.set_ylabel('$\log[M_{pred}$]', fontsize=14)
ax2.legend(fontsize=8, loc=4)

ax1.set_title('Power Law Predictions')

plt.tight_layout()
```

```python
f,ax = plt.subplots(figsize=[4,4])

ax.plot(one_to_one,one_to_one,'k',linestyle='dashed')
matt.binnedplot( sdm_dat[:,1],
            sdm_dat[:,0],
            n=50,
            percentiles = [35,47.5],
            ax=ax,
            label='sdm_',
            names=True,
            c='g',
            log=0
           )

ax.set_xlim(xmin=par['Y_min'], xmax=par['Y_max'])

ax.set_xlabel('$\log[M$]', fontsize=14)
ax.set_ylabel('$\log[M_{pred}$]', fontsize=14)
ax.legend(fontsize=8, loc=4)

ax.set_title('SDM Predictions')

plt.tight_layout()
```
